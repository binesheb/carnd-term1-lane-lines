## Create a gray Scale

```python
#To create a Gray Scale image we can use the function in CV2

gray = cv2.cvtColor(temp_img, cv2.COLOR_RGB2GRAY) #gray scale conversion
plt.imshow(gray, cmap='Greys_r')
plt.show()
```

## Process the Gray scale image

```python
# Canny Function is defined by cv2.Canny(blur_gray, low_threshold, high_threshold)

# we have to define values of blur_gray, low_threshold and high_threshold
# where blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size), 0)
# low_threshold defines the lower value of brightness,
# high_threshold defines the higher value of brightness
# kernel_size define the adjacency for creating blur function


kernel_size = 5
blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size), 0) #to suppress noise
plt.imshow(blur_gray , cmap='Greys_r')
plt.show()
print ('Figure : Blurred image to suppress any noise factor')
```

```python
#now we run canny edge detection function with low:high::1:3 value

low_threshold = 50
high_threshold = 150
edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
plt.imshow(edges)
plt.show()
```

```python
#The above result is in colour, we will convert to binary grayscale
plt.imshow(edges, cmap='Greys_r') #using cmap=Grey_r
plt.show()
```

```python
##Region of Interest
```

```python
### Create a Filter for shape area
```

```python
# we will create a mask layer having same dimension as original image
mask = np.zeros_like(edges) 
plt.imshow(mask)
plt.show()
print ('Dimension of Original :',edges.shape, ' and Dimension of Mask :', mask.shape)
```

```python
#definiton of Region of Interest : new way
```

```python
# since this is the same size as original image we can reuse xsize and ysize
vertices = np.array([[(0,ysize),(xsize*.45, ysize*.60), (xsize*.55, ysize*.60), (xsize,ysize)]], dtype=np.int32)


```

```python
ignore_mask_color = 255 
cv2.fillPoly(mask, vertices, ignore_mask_color)
masked_edges = cv2.bitwise_and(edges, mask)
plt.imshow(masked_edges)
plt.show()
```

```python
### Hough Transform
```

```python
# Hough Transform is defined by
#lines = cv2.HoughLinesP(image, rho, theta, threshold, np.array([]),min_line_length, max_line_gap)
# where, image
#rho
#theta 
#threshold > how many lines in same angle
#np.array
#min_line_length > number of pixel that make up a line
#max_line_gap > permissible blank pixel to combine as a single line

rho = 1
theta = np.pi/180
threshold = 5
min_line_length = 25
max_line_gap = 50
lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),min_line_length, max_line_gap)
```

```python
# Create a Blank Image
line_image = np.copy(image)*0
plt.imshow(line_image)
plt.show()
print ('Dimension of Original :',image.shape, ' and Dimension of Mask :', line_image.shape)
```

```python
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(0,255,0),8)
```

```python
color_edges = np.dstack((edges, edges, edges)) 
```

```python
lines_edges = cv2.addWeighted(color_edges, 0.8, line_image, 1, 0) 
plt.imshow(lines_edges)
plt.show()
```

```python
lines_edges = cv2.addWeighted(temp_img, .9, line_image, 1, 0) 
plt.imshow(lines_edges)
plt.show()
```

## Create a region of interest

```python
#lets start by defining vertices A,B,C as left_bottom, right_bottom and apex
left_bottom = (0,ysize)  #define lowest valueof X and max of Y
right_bottom = (xsize, ysize)
apex = (xsize*.5,ysize*.57)  
#the midpoint for Xvalue and the height is optimised for .57 of the actual image for optimised Field of View

#Now we can draw lines between these points using polyfit function.
#The function takes values of x1,x2,y1,y2 to make lines between (x1,y1) and (x2,y2)

fit_left = np.polyfit((left_bottom[0], apex[0]), (left_bottom[1], apex[1]), 1)
fit_right = np.polyfit((right_bottom[0], apex[0]), (right_bottom[1], apex[1]), 1)
fit_bottom = np.polyfit((left_bottom[0], right_bottom[0]), (left_bottom[1], right_bottom[1]), 1)
```

## Generation of Masking Area

```python
## Find the region inside the three lines

XX, YY = np.meshgrid(np.arange(0, xsize), np.arange(0, ysize))

region_thresholds = (YY > (XX*fit_left[0] + fit_left[1])) & \
                    (YY > (XX*fit_right[0] + fit_right[1])) & \
                    (YY < (XX*fit_bottom[0] + fit_bottom[1]))
        
plt.imshow(region_thresholds)
plt.show()
```

## Mask Overlay over the Actual Image

```python
#paint everything inside the region in RED [255,0,0]

mask_region=np.copy(temp_img) #A copy of the image is created to work upon with MASK Layer
mask_region[region_thresholds] = [255,0,0] # The selected region is painted RED 
plt.imshow(mask_region) #show the masked region
plt.show()
```

## Generation of Field of View under Observation under the masked region

```python
#paint everything inside the region in RED [255,0,0]

FOV=np.copy(temp_img) # create a copy
FOV[~region_thresholds] = [0,0,0] #mark everything not in the region to black
print ('The Final Field of View under Observation')
plt.imshow(FOV)
plt.show()
```

## Defining Addition Filter Parameter : Say Colour

#### We will define additional parameter to isolate the required info from the FOV image. 
#### Here we will use a colour filter to isolate the Bright lines from the Dark Road

```python
## Create a Filter Function to identify the Bright Colours
```

```python
#To define a Filter parameters, we create a array which holds all the parameters

red_threshold = 200
green_threshold = 200
blue_threshold = 200
rgb_threshold = [red_threshold, green_threshold, blue_threshold] 

#Now rgb_threshold can be used as a filter function to obtain the parameters
```

```python
## Identification of pixels which are under the threshold values
```

```python
# temp_img[:,:,0] represent RED Channel, 1 > GREEN , 2 > BLUE
org_img=np.copy(FOV) #we create a copy of FOV to preserve the original FOV file
color_threshold_img = (org_img[:,:,0] < rgb_threshold[0]) |  (org_img[:,:,1] < rgb_threshold[1]) | (org_img[:,:,2] < rgb_threshold[2])

plt.imshow(color_threshold_img)
plt.show()
```

```python
org_img[color_threshold_img] = [0,0,0]
plt.imshow(org_img)
plt.show()
```

```python
lane_img = np.copy(image)
lane_img[~color_threshold_img] = [255,0,0]
plt.imshow(lane_img)
plt.show()
```
